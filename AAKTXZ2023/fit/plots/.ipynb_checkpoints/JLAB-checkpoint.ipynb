```python
import os
import pandas as pd
import numpy as np
import pylab as py
import warnings
import matplotlib.pyplot as plt
from  matplotlib import rc
from matplotlib.colors import LogNorm
from matplotlib import font_manager
import matplotlib
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
rc('text',usetex=True)
from scipy.interpolate import interp1d
#from scipy.interpolate import make_interp_spline as spline
matplotlib.rcParams['text.latex.preamble']=[r"\usepackage{amsmath}"]
from matplotlib.ticker import ScalarFormatter,MaxNLocator,LogLocator,NullFormatter,FuncFormatter
warnings.filterwarnings('ignore')
import matplotlib.font_manager
from matplotlib.ticker import MultipleLocator, FormatStrFormatter,AutoMinorLocator
from matplotlib.pyplot import gca
```

```python
# Get Data 
path2dat = "../plot_data/"
datalst = os.listdir(path2dat)
```

```python
newdatalst = []
for address in datalst:
    if('JLAB' in address):
        newdatalst.append(address)
```

```python
df = pd.read_csv(path2dat+newdatalst[0], delimiter = ',',delim_whitespace=True)
```

```python
C_1 = df[0:10]
C_2 = df[10:20]
C_3 = df[20:30]
C_4 = df[30:40]
C_5 = df[40:50]
C_6 = df[50:60]
```

```python
Fe_1 = df[60:70]
Fe_2 = df[70:80]
Fe_3 = df[80:90]
Fe_4 = df[90:100]
Fe_5 = df[100:110]
Fe_6 = df[110:120]
```

```python
Pb_1 = df[120:130]
Pb_2 = df[130:140]
Pb_3 = df[140:150]
Pb_4 = df[150:160]
Pb_5 = df[160:170]
Pb_6 = df[170:180]
```

```python
def JLAB(axes,file,clr):    
    pt = (file["PTMIN"] + file["PTMAX"])/2
    Ra = file["MULT-RATIO"]
    err = file["ERR"]
    fuu = file['fuua']/file['fuu']/file['DIS']
    axes.errorbar(pt, Ra, yerr= err, fmt = "o",color = clr, ecolor = clr, uplims=False, lolims=False, capsize = 5, capthick =0.8, barsabove = False)
    axes.plot(pt,fuu,color = clr)
```

```python
fig, axs = plt.subplots(2, 3, sharex='col', sharey='row',
                        gridspec_kw={'hspace': 0, 'wspace': 0})

sizeOfFont = 25

JLAB(axs[0,0],C_1,"red")
JLAB(axs[0,0],Fe_1,"blue")
JLAB(axs[0,0],Pb_1,"black")


JLAB(axs[0,1],C_2,"red")
JLAB(axs[0,1],Fe_2,"blue")
JLAB(axs[0,1],Pb_2,"black")

JLAB(axs[0,2],C_3,"red")
JLAB(axs[0,2],Fe_3,"blue")
JLAB(axs[0,2],Pb_3,"black")


JLAB(axs[1,0],C_4,"red")
JLAB(axs[1,0],Fe_4,"blue")
JLAB(axs[1,0],Pb_4,"black")

JLAB(axs[1,1],C_5,"red")
JLAB(axs[1,1],Fe_5,"blue")
JLAB(axs[1,1],Pb_5,"black")

JLAB(axs[1,2],C_6,"red")
JLAB(axs[1,2],Fe_6,"blue")
JLAB(axs[1,2],Pb_6,"black")


axs[0,0].tick_params(axis = 'both', direction ='in', length = 10, labelsize = sizeOfFont)
axs[1,0].tick_params(axis = 'both', direction ='in', length = 10, labelsize = sizeOfFont)
axs[1,1].tick_params(axis = 'both', direction ='in', length = 10, labelsize = sizeOfFont)
axs[1,2].tick_params(axis = 'both', direction ='in', length = 10, labelsize = sizeOfFont)


axs[0,0].set_xscale("log")
axs[0,1].set_xscale("log")
axs[0,2].set_xscale("log")

fig.set_size_inches(20,10)
```
