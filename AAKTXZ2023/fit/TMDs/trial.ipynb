```python
df_like_H  = pd.read_csv("like_H.csv",delim_whitespace=True)
df_like_Ne = pd.read_csv("like_Ne.csv",delim_whitespace=True)
df_like_Kr = pd.read_csv("like_Kr.csv",delim_whitespace=True)
df_like_Xe = pd.read_csv("like_Xe.csv",delim_whitespace=True)
```

```python
df_new_H  = pd.read_csv("new_H.csv",delim_whitespace=True)
df_new_Ne = pd.read_csv("new_Ne.csv",delim_whitespace=True)
df_new_Kr = pd.read_csv("new_Kr.csv",delim_whitespace=True)
df_new_Xe = pd.read_csv("new_Xe.csv",delim_whitespace=True)
```

```python
fig, axes = plt.subplots(nrows=7, ncols=4, sharex='col', sharey='row',
                        gridspec_kw={'hspace': 0, 'wspace': 0})

#### All share the same axis (plot)
######## Set Size of Font
SizeOfFont = 22

### Set Figure Size
fig.set_size_inches(18,12)


df   = df_new_H
df2  = df_like_H
i=0
axes[0,i].plot(df['Z'], df['U']/df2['U'],   color = 'black', linewidth = 2.0)
axes[1,i].plot(df['Z'], df['UB']/df2['UB'], color = 'black', linewidth = 2.0)
axes[2,i].plot(df['Z'], df['D']/df2['D'],   color = 'black', linewidth = 2.0)
axes[3,i].plot(df['Z'], df['DB']/df2['DB'], color = 'black', linewidth = 2.0)
axes[4,i].plot(df['Z'], df['S']/df2['S'],   color = 'black', linewidth = 2.0)
axes[5,i].plot(df['Z'], df['SB']/df2['SB'], color = 'black', linewidth = 2.0)
axes[6,i].plot(df['Z'], df['GL']/df2['GL'], color = 'black', linewidth = 2.0)



df   = df_new_Ne
df2  = df_like_Ne
i=1
axes[0,i].plot(df['Z'], df['U']/df2['U'],   color = 'black', linewidth = 2.0)
axes[1,i].plot(df['Z'], df['UB']/df2['UB'], color = 'black', linewidth = 2.0)
axes[2,i].plot(df['Z'], df['D']/df2['D'],   color = 'black', linewidth = 2.0)
axes[3,i].plot(df['Z'], df['DB']/df2['DB'], color = 'black', linewidth = 2.0)
axes[4,i].plot(df['Z'], df['S']/df2['S'],   color = 'black', linewidth = 2.0)
axes[5,i].plot(df['Z'], df['SB']/df2['SB'], color = 'black', linewidth = 2.0)
axes[6,i].plot(df['Z'], df['GL']/df2['GL'], color = 'black', linewidth = 2.0)



df   = df_new_Kr
df2  = df_like_Kr
i=2
axes[0,i].plot(df['Z'], df['U']/df2['U'],   color = 'black', linewidth = 2.0)
axes[1,i].plot(df['Z'], df['UB']/df2['UB'], color = 'black', linewidth = 2.0)
axes[2,i].plot(df['Z'], df['D']/df2['D'],   color = 'black', linewidth = 2.0)
axes[3,i].plot(df['Z'], df['DB']/df2['DB'], color = 'black', linewidth = 2.0)
axes[4,i].plot(df['Z'], df['S']/df2['S'],   color = 'black', linewidth = 2.0)
axes[5,i].plot(df['Z'], df['SB']/df2['SB'], color = 'black', linewidth = 2.0)
axes[6,i].plot(df['Z'], df['GL']/df2['GL'], color = 'black', linewidth = 2.0)



df   = df_new_Xe
df2  = df_like_Xe
i=3
axes[0,i].plot(df['Z'], df['U']/df2['U'],   color = 'black', linewidth = 2.0)
axes[1,i].plot(df['Z'], df['UB']/df2['UB'], color = 'black', linewidth = 2.0)
axes[2,i].plot(df['Z'], df['D']/df2['D'],   color = 'black', linewidth = 2.0)
axes[3,i].plot(df['Z'], df['DB']/df2['DB'], color = 'black', linewidth = 2.0)
axes[4,i].plot(df['Z'], df['S']/df2['S'],   color = 'black', linewidth = 2.0)
axes[5,i].plot(df['Z'], df['SB']/df2['SB'], color = 'black', linewidth = 2.0)
axes[6,i].plot(df['Z'], df['GL']/df2['GL'], color = 'black', linewidth = 2.0)




    
##### Axis Limits
for i in range(7):
    axes[i,0].set_ylim(0.8,1.2)
    axes[i,0].set_yticks([0.9,1.0,1.1])


############## Axis Labels
distanceFromAxis = 45
axes[0,0].set_ylabel(r'\rm $\frac{D_{\rm{\pi^+}/u}^{NewFF}(z)}{D_{\rm{\pi^+}/u}^{LIKEn}(z)}$',       fontsize = SizeOfFont, rotation = 0, labelpad = distanceFromAxis )
axes[1,0].set_ylabel(r'\rm $\frac{D_{\rm{\pi^+}/\bar{u}}^{NewFF}(z)}{D_{\rm{\pi^+}/\bar{u}}^{LIKEn}(z)}$', fontsize = SizeOfFont, rotation = 0, labelpad = distanceFromAxis )
axes[2,0].set_ylabel(r'\rm $\frac{D_{\rm{\pi^+}/d}^{NewFF}(z)}{D_{\rm{\pi^+}/d}^{LIKEn}(z)}$',  fontsize = SizeOfFont, rotation = 0, labelpad = distanceFromAxis )
axes[3,0].set_ylabel(r'\rm $\frac{D_{\rm{\pi^+}/\bar{d}}^{NewFF}(z)}{D_{\rm{\pi^+}/\bar{d}}^{LIKEn}(z)}$', fontsize = SizeOfFont, rotation = 0, labelpad = distanceFromAxis )
axes[4,0].set_ylabel(r'\rm $\frac{D_{\rm{\pi^+}/s}^{NewFF}(z)}{D_{\rm{\pi^+}/s}^{LIKEn}(z)}$', fontsize = SizeOfFont, rotation = 0, labelpad = distanceFromAxis )
axes[5,0].set_ylabel(r'\rm $\frac{D_{\rm{\pi^+}/\bar{s}}^{NewFF}(z)}{D_{\rm{\pi^+}/\bar{s}}^{LIKEn}(z)}$', fontsize = SizeOfFont, rotation = 0, labelpad = distanceFromAxis )
axes[6,0].set_ylabel(r'\rm $\frac{D_{\rm{\pi^+}/g}^{NewFF}(z)}{D_{\rm{\pi^+}/\bar{g}}^{LIKEn}(z)}$', fontsize = SizeOfFont, rotation = 0, labelpad = distanceFromAxis )

#axes[5,1].set_xlabel(r"\rm $z$", fontsize = SizeOfFont,rotation = 0,labelpad = 15)
fig.text(0.5, 0, r"\rm $z$", ha='center', va='center',fontsize = SizeOfFont,rotation = 0)


############## Axis Ticks
for i in range(7):
    for j in range(4):
        axes[i,j].tick_params(axis = 'both', direction ='in',labelsize = SizeOfFont)

for i in range(4):
  axes[3,i].set_xticks([0.2, 0.4, 0.6])
  axes[3,i].set_xlim([0.05,0.7])


#################### Code for Textboxes 
props = dict(boxstyle='round', facecolor='gray', alpha=0)
axes[0,0].text(0.25, 0.95, r"\rm Medium = Vacuum "          ,transform=axes[0,0].transAxes, fontsize=SizeOfFont,
    verticalalignment='top', bbox=props)
axes[0,1].text(0.50, 0.95, r"\rm $\rm{Ne}$ "         ,transform=axes[0,1].transAxes, fontsize=SizeOfFont,
    verticalalignment='top', bbox=props) 
axes[0,2].text(0.50, 0.95, r"\rm $\rm{Kr}$ "               ,transform=axes[0,2].transAxes, fontsize=SizeOfFont,
    verticalalignment='top', bbox=props)
axes[0,3].text(0.50, 0.95, r"\rm $\rm{Xe}$ "         ,transform=axes[0,3].transAxes, fontsize=SizeOfFont,
    verticalalignment='top', bbox=props)




################# Legends
#axes[1,3].legend(frameon = False,fontsize = SizeOfFont,loc='upper center', bbox_to_anchor=(0.6, 0.9))

##### Save Figure
plt.tight_layout()
plt.show()
fig.savefig('Q_1_new.pdf', bbox_inches = "tight")
```
