```python
class pdfdata:
    def __init__(self, file):
        self.x   = file['x']
        self.ut  = file['ut']
        self.dt  = file['dt']
        self.st  = file['st']
        self.ubt = file['ubt']
        self.dbt = file['dbt']
        self.sbt = file['sbt']
        self.glt = file['glt']
        self.u   = file['u']
        self.d   = file['d']
        self.s   = file['s']
        self.ub  = file['ub']
        self.db  = file['db']
        self.sb  = file['sb']
        self.gl  = file['gl']
```

```python
def pdf_plot(data,axes):
    axes.plot(data.x, data.ut, color = 'blue', label = r'\rm $u$', alpha =2)
    axes.plot(data.x, data.dt, color = 'red', label = r'\rm $d$', alpha =1)
    axes.plot(data.x, data.st, color = 'green', label = r'\rm $s$', alpha =1)
    axes.plot(data.x, data.ubt, color = 'darkblue', label = r'\rm $\bar{u}$', alpha =1)
    axes.plot(data.x, data.dbt, color = 'darkred', label = r'\rm $\bar{d}$', alpha =1)
    axes.plot(data.x, data.sbt, color = 'darkgreen', label = r'\rm $\bar{s}$', alpha =1)
    axes.plot(data.x, data.glt, color = 'black', label = r'\rm $g$', alpha =1)
    
    axes.plot(data.x, data.u, color = 'blue', linestyle = 'dashed')
    axes.plot(data.x, data.d, color = 'red', linestyle = 'dashed')
    axes.plot(data.x, data.s, color = 'green', linestyle = 'dashed')
    axes.plot(data.x, data.ub, color = 'darkblue', linestyle = 'dashed')
    axes.plot(data.x, data.db, color = 'darkred', linestyle = 'dashed')
    axes.plot(data.x, data.sb, color = 'darkgreen', linestyle = 'dashed')
    axes.plot(data.x, data.gl, color = 'black', linestyle = 'dashed')
    axes.tick_params(axis = 'both', direction ='in',labelsize = 25)
```

```python
def pdf_plot_yellow(data,axes):
    axes.plot(data.x, data.ut, color = 'blue', label = r'\rm $u$', alpha =2)
    axes.plot(data.x, data.dt, color = 'red', label = r'\rm $d$', alpha =1)
    axes.plot(data.x, data.st, color = 'green', label = r'\rm $s$', alpha =1)
    axes.plot(data.x, data.ubt, color = 'darkblue', label = r'\rm $\bar{u}$', alpha =1)
    axes.plot(data.x, data.dbt, color = 'darkred', label = r'\rm $\bar{d}$', alpha =1)
    axes.plot(data.x, data.sbt, color = 'darkgreen', label = r'\rm $\bar{s}$', alpha =1)
    axes.plot(data.x, data.glt, color = 'black', label = r'\rm $g$', alpha =1)
    
    axes.plot(data.x, data.u, color = 'yellow', linestyle = 'dashed', alpha = 1)
    axes.plot(data.x, data.d, color = 'yellow', linestyle = 'dashed', alpha = 1)
    axes.plot(data.x, data.s, color = 'yellow', linestyle = 'dashed', alpha = 1)
    axes.plot(data.x, data.ub, color = 'yellow', linestyle = 'dashed', alpha = 1)
    axes.plot(data.x, data.db, color = 'yellow', linestyle = 'dashed', alpha = 1)
    axes.plot(data.x, data.sb, color = 'yellow', linestyle = 'dashed', alpha = 1)
    axes.plot(data.x, data.gl, color = 'yellow', linestyle = 'dashed', alpha = 1)
    axes.tick_params(axis = 'both', direction ='in',labelsize = 25)
```

```python
def pdf_plott(data,axes):
    axes.plot(data.x, data.u, color = 'blue', linestyle = 'dashed')
    axes.plot(data.x, data.d, color = 'red', linestyle = 'dashed')
    axes.plot(data.x, data.s, color = 'green', linestyle = 'dashed')
    axes.plot(data.x, data.ub, color = 'darkblue', linestyle = 'dashed')
    axes.plot(data.x, data.db, color = 'darkred', linestyle = 'dashed')
    axes.plot(data.x, data.sb, color = 'darkgreen', linestyle = 'dashed')
    axes.plot(data.x, data.gl, color = 'black', linestyle = 'dashed')
    axes.tick_params(axis = 'both', direction ='in',labelsize = 25)
```

```python
def plotfile(file,axes):
    data = pdfdata(file)
    pdf_plot(data,axes)

def plotfilecomp(file,axes):
    data = pdfdata(file)
    pdf_plott(data,axes)
    
def plotfileyellow(file,axes):
    data = pdfdata(file)
    pdf_plot_yellow(data,axes)
```

```python
fig,axs = plt.subplots(4, 4, sharex='col', sharey='row', gridspec_kw={'hspace': 0, 'wspace': 0})

plotfile(df[0],axs[0,0])
plotfile(df[1],axs[1,0])
plotfile(df[2],axs[2,0])
plotfile(df[3],axs[3,0])

plotfile(df[4],axs[0,1])
plotfile(df[5],axs[1,1])
plotfile(df[6],axs[2,1])
plotfile(df[7],axs[3,1])

plotfile(df[8],axs[0,2])
plotfile(df[9],axs[1,2])
plotfile(df[10],axs[2,2])
plotfile(df[11],axs[3,2])

plotfile(df[12],axs[0,3])
plotfile(df[13],axs[1,3])
plotfile(df[14],axs[2,3])
plotfile(df[15],axs[3,3])

# Axes Limits
# Y axis
axs[0,0].set_ylim(0.01,1)
axs[1,0].set_ylim(0.01,1)
axs[2,0].set_ylim(0.01,1)
axs[3,0].set_ylim(0,1)

#X axis
axs[3,0].set_xscale("log")
axs[3,1].set_xscale("log")
axs[3,2].set_xscale("log")
axs[3,3].set_xscale("log")


axs[0,0].set_xlim(1e-3,1e0)
axs[0,1].set_xlim(1.01e-3,1e0)
axs[0,2].set_xlim(1.01e-3,1e0)
axs[0,3].set_xlim(1.01e-3,1e0)


# Set Labels
axs[0,0].set_ylabel(r"\rm $xf_p(x,Q^2)$", fontsize=40,rotation = 0,labelpad = 90)
axs[3,1].set_xlabel(r"\rm $x$", fontsize=40)


#Props
props = dict(boxstyle='round', facecolor='white', alpha=0)
props2 = dict(boxstyle='round', facecolor='gray', alpha=0.4)

axs[0,0].text(0.2, 1.18,  r'\rm $Q = 2$ GeV', transform=axs[0,0].transAxes, fontsize=40, color = 'black',
verticalalignment='top', bbox=props2)

axs[0,1].text(0.2, 1.18,  r'\rm $Q = 5$ GeV', transform=axs[0,1].transAxes, fontsize=40, color = 'black',
verticalalignment='top', bbox=props2)

axs[0,2].text(0.2, 1.18,  r'\rm $Q = 10$ GeV', transform=axs[0,2].transAxes, fontsize=40, color = 'black',
verticalalignment='top', bbox=props2)

axs[0,3].text(0.2, 1.18,  r'\rm $Q = 50$ GeV', transform=axs[0,3].transAxes, fontsize=40, color = 'black',
verticalalignment='top', bbox=props2)

axs[0,0].text(0.2, 0.8,  r'\rm He', transform=axs[0,0].transAxes, fontsize=40, color = 'black',
verticalalignment='top', bbox=props)

axs[1,0].text(0.2, 0.8,  r'\rm Ne', transform=axs[1,0].transAxes, fontsize=40, color = 'black',
verticalalignment='top', bbox=props)

axs[2,0].text(0.2, 0.8,  r'\rm Kr', transform=axs[2,0].transAxes, fontsize=40, color = 'black',
verticalalignment='top', bbox=props)

axs[3,0].text(0.2, 0.8,  r'\rm Xe', transform=axs[3,0].transAxes, fontsize=40, color = 'black',
verticalalignment='top', bbox=props)

fig.set_size_inches(25,20)
axs[0,3].legend(frameon = False,fontsize=40,loc='upper center', bbox_to_anchor=(1.3, 0.8))
fig.savefig('PDFs_LHAPDF.pdf', bbox_inches = "tight")    
```
