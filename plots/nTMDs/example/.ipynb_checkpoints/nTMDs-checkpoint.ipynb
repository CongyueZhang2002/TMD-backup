```python
def interpolate_axes(axes,xdata,ydata,ccolor,isLabeled,llabel):
    end = len(xdata)
    xxdata = xdata[1:end]
    yydata = ydata[1:end]
    xmin = np.min(xxdata)
    xmax = np.max(xxdata)
    xnew = np.linspace(xmin,xmax,num = 60, endpoint = True)
    ynew = interp1d(xxdata,yydata, kind = 'cubic')
    if(isLabeled):
        axes.plot(xnew,ynew(xnew), color = ccolor, label = llabel )
    else:
        axes.plot(xnew,ynew(xnew), color = ccolor)
```

```python
fig1, axs1 = plt.subplots(2, 3, sharex='col', sharey='row', gridspec_kw={'hspace': 0, 'wspace': 0})

interpolate_axes(axs1[0,0], df1['kt'], df1['up'], 'blue', False,"" )
interpolate_axes(axs1[0,1], df1['kt'], df1['u1'], 'blue', False,"" )
interpolate_axes(axs1[0,2], df1['kt'], df1['u2'], 'blue', True,r"\rm $u$")


# Plot Labels
axs1[0,0].set_ylabel(r"\rm $f(x,k_t)$", fontsize = 25,rotation = 0,labelpad = 50)
axs1[0,1].set_xlabel(r"\rm $k_t$ (GeV)", fontsize = 25, labelpad = 10)

axs1[0,0].tick_params(axis = 'both', direction ='in',labelsize = 25)
axs1[0,1].tick_params(axis = 'both', direction ='in',labelsize = 25)
axs1[0,2].tick_params(axis = 'both', direction ='in',labelsize = 25)

# Plot Limits
axs1[0,0].set_xlim(0.01,1)
axs1[0,1].set_xlim(0.01,1)
axs1[0,2].set_xlim(0.01,1)


interpolate_axes(axs1[1,0], df2['kt'], df2['up'], 'blue', False,"" )
interpolate_axes(axs1[1,1], df2['kt'], df2['u1'], 'blue', False,"" )
interpolate_axes(axs1[1,2], df2['kt'], df2['u2'], 'blue', True,r"\rm $u$")



# Plot Labels

axs1[1,0].set_ylabel(r"\rm $f(x,k_T)$", fontsize = 25,rotation = 0,labelpad = 50)
axs1[1,1].set_xlabel(r"\rm $k_T$ (GeV)", fontsize = 25, labelpad = 10)

axs1[1,0].tick_params(axis = 'both', direction ='in',labelsize = 25)
axs1[1,1].tick_params(axis = 'both', direction ='in',labelsize = 25)
axs1[1,2].tick_params(axis = 'both', direction ='in',labelsize = 25)

# Plot Limits
axs1[1,0].set_xlim(0.01,1)
axs1[1,1].set_xlim(0.01,1)
axs1[1,2].set_xlim(0.01,1)

# Text Boxes ##
props = dict(boxstyle='round', facecolor='white', alpha=0)

props2 = dict(boxstyle='round', facecolor='grey', alpha=0.2)


axs1[0,0].text(0.2, 0.2,  r"\rm $Q = Q_0$", transform=axs1[0,0].transAxes, fontsize=30,
verticalalignment='top', bbox=props2)

axs1[1,0].text(0.2, 0.2,  r"\rm $Q = 10$ GeV", transform=axs1[1,0].transAxes, fontsize=30,
verticalalignment='top', bbox=props2)


axs1[0,0].text(0.7, 0.9,  r"\rm $p/n$", transform=axs1[0,0].transAxes, fontsize=30,
verticalalignment='top', bbox=props)

axs1[0,1].text(0.7, 0.9,  r"\rm He", transform=axs1[0,1].transAxes, fontsize=30,
verticalalignment='top', bbox=props)

axs1[0,2].text(0.7, 0.9,  r"\rm Xe", transform=axs1[0,2].transAxes, fontsize=30,
verticalalignment='top', bbox=props)

axs1[0,2].legend(frameon = False,fontsize = 25,loc='upper center', bbox_to_anchor=(1.2, 0.9))

fig1.set_size_inches(20,12)
```
