```python
for _fp in newdatalst:
    print(_fp)
    maxlib[_fp] = []
    minlib[_fp] = []
    barlib[_fp] = []
    if('pt2' in _fp):
        rt = '/data_reps/data_rep_000/HERMES/'
    else:
        rt = '/data_reps/data_rep_000/'
    for j in range(len(pd.read_csv(path2dat+rt+_fp, delimiter = ',',delim_whitespace=True))):
        for rep in dfinit:
            rep = int(rep)
            if rep == 0:
                fp0 = path2dat+rt+_fp
                df = pd.read_csv(fp0, delimiter = ',',delim_whitespace=True)[j:j+1]
            else:
                if len(str(rep)) == 1:
                    numb = '00'+str(rep)
                elif len(str(rep)) == 2:
                    numb = '0'+str(rep)
                else:
                    numb = str(rep)
                if('pt2' in _fp):    
                    fp1 = path2dat+'/data_reps/data_rep_'+numb+'/HERMES/'+_fp
                else:
                    fp1 = path2dat+'/data_reps/data_rep_'+numb+'/'+_fp
                df1 = pd.read_csv(fp1, delimiter = ',',delim_whitespace=True)[j:j+1]
                df = pd.concat([df,df1])
        if 'pi' in _fp:
                ratio = df['fuua']/df['fuu']/df['DIS']
                sortedlst = sorted(ratio.tolist())
                downcutoff = int(len(sortedlst)/100.*15)
                upcutoff = int(len(sortedlst)/100.*84)
                maxlib[_fp].append(sortedlst[upcutoff])
                minlib[_fp].append(sortedlst[downcutoff])
                barlib[_fp].append(ratio.iloc[0])
        else:
                ratio = df['R']
                sortedlst = sorted(ratio.tolist())
                downcutoff = int(len(sortedlst)/100.*15)
                upcutoff = int(len(sortedlst)/100.*84)
                maxlib[_fp].append(sortedlst[upcutoff])
                minlib[_fp].append(sortedlst[downcutoff])
                barlib[_fp].append(ratio.iloc[0])
```

```python
errsqu = {}
errsqb = {}
for _fp in newdatalst:
    if('pt2' in _fp):
        rt = '/data_reps_col/data_rep_000/HERMES/'
    else:
        rt = '/data_reps_col/data_rep_000/'
    fp0 = path2dat+ rt+ _fp
    df0 = pd.read_csv(fp0, delimiter = ',',delim_whitespace=True)
    errsqu[_fp] = []
    errsqb[_fp] = []
    #print(len(df0),_fp)
    for i in range(len(df0)):
        errsqu[_fp].append(0)
        errsqb[_fp].append(0)
    for i in range(55):
        j = 2*(i+1)
        Splus = j
        Sminus = j-1
        #if (_fp == 'fort.1'):
        #    print(Sminus,Splus)
        if len(str(Splus)) == 1:
            numbplus = '00'+str(Splus)
        elif len(str(Splus)) == 2:
            numbplus = '0'+str(Splus)
        else:
            numbplus = str(Splus)
        if len(str(Sminus)) == 1:
            numbminus = '00'+str(Sminus)
        elif len(str(Sminus)) == 2:
            numbminus = '0'+str(Sminus)
        else:
            numbminus = str(Sminus)
        if('pt2' in _fp):
            fpp = path2dat+'data_reps_col/data_rep_'+numbplus+'/HERMES/'+_fp
            fpm = path2dat+'data_reps_col/data_rep_'+numbminus+'/HERMES/'+_fp
        else:
            fpp = path2dat+'data_reps_col/data_rep_'+numbplus+'/'+_fp
            fpm = path2dat+'data_reps_col/data_rep_'+numbminus+'/'+_fp 
        dfp = pd.read_csv(fpp, delimiter = ',',delim_whitespace=True)
        dfm = pd.read_csv(fpm, delimiter = ',',delim_whitespace=True)
        if('pi' in _fp):
                 ratiop = dfp['fuua']/dfp['fuu']/dfp['DIS']
                 ratio0 = df0['fuua']/df0['fuu']/df0['DIS']
                 ratiom = dfm['fuua']/dfm['fuu']/dfm['DIS']
        else:
                ratiop = dfp['R']
                ratio0 = df0['R']
                ratiom = dfm['R']
        for k in range(len(dfp)):
                errp = ratiop[k]-ratio0[k]
                errm = ratiom[k]-ratio0[k]
                maxx = np.max([errp,errm,0])
                minn = np.min([errp,errm,0])
                if(i < 48):
                     errsqu[_fp][k] = errsqu[_fp][k] + (maxx/1.645)**2
                     errsqb[_fp][k] = errsqb[_fp][k] + (minn/1.645)**2
                else:
                     errsqu[_fp][k] = errsqu[_fp][k] + (maxx)**2
                     errsqb[_fp][k] = errsqb[_fp][k] + (minn)**2
```

```python
maxrat = {}
minrat = {}
for _fp in newdatalst:
    maxrat[_fp] = []
    minrat[_fp] = []
    #print(_fp)
    if('pt2' in _fp):
            fp0 = path2dat+'/data_reps/data_rep_000/HERMES/'+ _fp
    else:
            fp0 = path2dat+'/data_reps/data_rep_000/'+ _fp
    df0 = pd.read_csv(fp0, delimiter = ',',delim_whitespace=True)
    if('pi' in _fp):
            centralrat = df0['fuua']/df0['fuu']/df0['DIS']
    else:
            centralrat = df0['R']
    for i in range(len(centralrat)):
        ERRU = np.sqrt ( errsqu[_fp][i] +  ((maxlib[_fp][i]-minlib[_fp][i])/2)**2 )
        ERRB = np.sqrt ( errsqb[_fp][i] +  (( maxlib[_fp][i]-minlib[_fp][i])/2)**2 ) 
        maxrat[_fp].append(centralrat[i] + ERRU)
        minrat[_fp].append(centralrat[i] - ERRB)
```

```python
def pred(axes,data, clr, lbl, factor, zord = 1):
    rt = '/data_reps/data_rep_000/'
    file = pd.read_csv(path2dat + rt + data, delimiter = ',',delim_whitespace=True)
    pt = file['PhT']
    maxx = maxlib[data]
    minn = minlib[data]
    maxxc = maxrat[data]
    minnc = minrat[data]
    pt_new = np.linspace(0.05,0.65, 200)
    f = CubicSpline(pt[0:9], maxx[0:9], bc_type='natural')
    maxx_new = f(pt_new)
    f = CubicSpline(pt[0:9], minn[0:9], bc_type='natural')
    minn_new = f(pt_new)
    f = CubicSpline(pt[0:9], maxxc[0:9], bc_type='natural')
    maxxc_new = f(pt_new)
    f = CubicSpline(pt[0:9], minnc[0:9], bc_type='natural')
    minnc_new = f(pt_new)
    axes.fill_between(pt_new,maxx_new +factor,minn_new+factor,color = clr,alpha=1,label=lbl,zorder = zord)
    if(collinear == 1):
        axes.fill_between(pt_new,maxxc_new+ factor,minnc_new+factor,color = clr,alpha=0.3, zorder = zord)
```

```python
fig, ax = plt.subplots(1, 1, sharex='col', sharey='row',
                        gridspec_kw={'hspace': 0, 'wspace': 0})

sizeOfFont =20
collinear = 1
pred(ax, 'EIC_Qhi.dat','red', r"\rm EIC, $Q^2 = 100$ GeV$^2$, $x = 0.05$, $A = $ Au",0.,zord = 0)
pred(ax, 'EIC_Qlo.dat','black', r"\rm EIC , $Q^2 = 4$ GeV$^2$, $x = 0.05$, $A =$ Au",0.)
pred(ax, 'JLAB_pred.dat','green', r"\rm JLab, $Q^2 = 2.5$ GeV$^2$, $x = 0.4$, $A = $ Pb",0)
#pred(ax, 'EIC_Qmed.dat','blue', "Qmed",1.1)


ax.set_xlabel(r'\rm $P_{h\perp}$ (GeV)',  fontsize = sizeOfFont,labelpad = 10)
ax.set_ylabel(r'\rm $R_{A}^{\pi^+}$', fontsize = sizeOfFont,labelpad = 10)
ax.tick_params(axis = 'both', direction ='in', length = 10, labelsize = sizeOfFont)


ax.tick_params(axis='both',which = "major",direction = "in",length = 7)
ax.minorticks_on()
ax.yaxis.set_ticks_position('both')
ax.xaxis.set_ticks_position('both')
ax.tick_params(axis='both',which = "minor",direction = "in",length = 4)


ax.legend(frameon = False,fontsize = sizeOfFont-6,loc='upper left')

ax.set_xlim(0,0.7)
ax.set_ylim(0.50,0.9)
ax.set_xticks([0.2,0.4,0.6])


fig.set_size_inches(6,5)


fig.savefig('EIC_JLAB.pdf', bbox_inches = "tight")

```

```python
def HERMES(axes, filename, clr, row, kinematics, lbl,ffmt):
    path2dat = pwd+'/data_reps/data_rep_000/HERMES/'
    if(kinematics == 'Nu'):
        filenames = filename + '_nu.dat'
    elif(kinematics == 'Z'):
        filenames = filename + '_z.dat'
    elif(kinematics == 'q2'):
        filenames = filename + '_q2.dat'
    elif(kinematics == 'pt2'):
        filenames = filename + '_pt2.dat'
    if (filename != 'pip_kr_pt2.dat'):
            file = (pd.read_csv(path2dat + filenames, sep ='\s+'))
            maxx = maxlib[filenames]
            minn = minlib[filenames]
            ydata = file['MULT-RATIO']
            error = np.sqrt((file['STAT'])**2+ (file['SYS'])**2)
            xdata = np.sqrt(file[kinematics])
            axes.errorbar(xdata, ydata, error, fmt = ffmt,color = clr,label = lbl,capsize = 5)
            axes.step(xdata,maxx, where = 'mid',color = clr,alpha=0.4)
            axes.step(xdata,minn, where = 'mid',color = clr,alpha=0.4)
            axes.fill_between(xdata,maxx,minn, step = 'mid',color = clr,alpha=0.4)
            
            
    if(collinear == 1):
                path2dat2 = pwd+'/data_reps_col/data_rep_000/HERMES/'
                maxx = maxrat[filenames]
                minn = minrat[filenames]
                #axes.step(xdata,maxx, where = 'mid',color = clr,alpha=0.2)
                #axes.step(xdata,minn, where = 'mid',color = clr,alpha=0.2)
                axes.fill_between(xdata,maxx,minn, step = 'mid',color = clr,alpha=0.2)
```

```python
fig, axs = plt.subplots(1, 1, sharex='col', sharey='row',gridspec_kw={'hspace': 0, 'wspace': 0})

sizeOfFont = 22

collinear = 1


HERMES(axs, 'pi0_he', 'red', 0, 'pt2', r'\rm He',"o")
HERMES(axs, 'pi0_ne', 'black', 0, 'pt2', r'\rm Ne',"^")
HERMES(axs, 'pi0_kr', 'green', 0, 'pt2', r'\rm Kr',"v")
HERMES(axs, 'pi0_xe', 'blue', 0, 'pt2', r'\rm Xe',"*")


axs.tick_params(axis="both",direction="in", length=10,labelsize = sizeOfFont)

axs.set_ylim(0.40,1.15)

axs.set_xlim(0.12,0.55)

#axs.set_xticks([0.0,0.1,0.2])
axs.set_yticks([0.6,0.8,1.0] )


axs.set_xlabel(r'\rm $P_{h\perp}$ (GeV)',fontsize=sizeOfFont)
axs.set_ylabel(r"\rm $R^h_A$", fontsize = sizeOfFont)
               

axs.legend(frameon = False,fontsize = sizeOfFont-7,loc='upper right',ncol = 1,columnspacing = 0.5)

    
props = dict(boxstyle='round', facecolor='white', alpha=0)

axs.text(0.1, 0.20,  r"\rm $ h = \pi^0$", transform=axs.transAxes, fontsize=sizeOfFont,
verticalalignment='top', bbox=props)

plt.show()

fig.savefig('HERMES_pt2.pdf', bbox_inches = "tight")
```

```python
def JLAB(axes, nucleus, binn, clr, lbl, hadron = 'pi+'):
    if(hadron == 'pi+'):
        if(nucleus == 'C'):
            file = C[binn]
            maxx = Cmax[binn]
            minn = Cmin[binn]
            maxxc = Cmaxr[binn]
            minnc = Cminr[binn]
        elif(nucleus == 'Fe'):
            file = Fe[binn]
            maxx = Femax[binn]
            minn = Femin[binn]
            maxxc = Femaxr[binn]
            minnc = Feminr[binn]
        elif(nucleus == 'Pb'):
            file = Pb[binn]
            maxx = Pbmax[binn]
            minn = Pbmin[binn]
            maxxc = Pbmaxr[binn]
            minnc = Pbminr[binn]
    elif(hadron == 'pi-'):
        if(nucleus == 'C'):
            file = Cm[binn]
            maxx = Cmaxm[binn]
            minn = Cminm[binn]
            maxxc = Cmaxrm[binn]
            minnc = Cminrm[binn]
        elif(nucleus == 'Fe'):
            file = Fem[binn]
            maxx = Femaxm[binn]
            minn = Feminm[binn]
            maxxc = Femaxrm[binn]
            minnc = Feminrm[binn]
        elif(nucleus == 'Pb'):
            file = Pbm[binn]
            maxx = Pbmaxm[binn]
            minn = Pbminm[binn]
            maxxc = Pbmaxrm[binn]
            minnc = Pbminrm[binn]
    pt = (file["PTMIN"] + file["PTMAX"])/2
    Ra = file["MULT-RATIO"]
    err = file["ERR"]
    fuu = file['fuua']/file['fuu']/file['DIS']
    axes.step(pt,maxx, where = 'mid',color = clr,alpha=0.4)
    axes.step(pt,minn, where = 'mid',color = clr,alpha=0.4)
    axes.fill_between(pt,maxx,minn, step = 'mid',color = clr,alpha=0.4)
    axes.errorbar(pt, Ra, yerr= err, fmt = "o",color = clr, ecolor = clr, uplims=False, lolims=False, capsize = 5, capthick =0.8, barsabove = False,label=lbl)
    if(collinear == 1):
        axes.fill_between(pt,maxxc,minnc, step = 'mid',color = clr,alpha=0.2)
```

```python
def JLAB2(axes, nucleus, binn, clr, lbl, factor =1.0):
    if(nucleus == 'C'):
        file = C[binn]
        maxx = Cmax[binn]
        minn = Cmin[binn]
        maxxc = Cmaxr[binn]
        minnc = Cminr[binn]
    elif(nucleus == 'Fe'):
        file = Fe[binn]
        maxx = Femax[binn]
        minn = Femin[binn]
        maxxc = Femaxr[binn]
        minnc = Feminr[binn]
    elif(nucleus == 'Pb'):
        file = Pb[binn]
        maxx = Pbmax[binn]
        minn = Pbmin[binn]
        maxxc = Pbmaxr[binn]
        minnc = Pbminr[binn]
    pt = (file["PTMIN"] + file["PTMAX"])/2
    Ra = file["MULT-RATIO"]
    err = file["ERR"]
    fuu = file['fuua']/file['fuu']/file['DIS']
    #axes.step(pt,maxx, where = 'mid',color = clr,alpha=0.4)
    #axes.step(pt,minn, where = 'mid',color = clr,alpha=0.4)
    
    f = CubicSpline(pt[0:8], maxx[0:8], bc_type='natural')
    pt_new = np.linspace(0.05,0.70, 200)
    maxx_new = f(pt_new)
    f = CubicSpline(pt[0:8], minn[0:8], bc_type='natural')
    minn_new = f(pt_new)
    f = CubicSpline(pt[0:8], maxxc[0:8], bc_type='natural')
    maxxc_new = f(pt_new)
    f = CubicSpline(pt[0:8], minnc[0:8], bc_type='natural')
    minnc_new = f(pt_new)
    axes.fill_between(pt_new,factor*maxx_new,factor*minn_new,color = clr,alpha=0.4,label= lbl)
    #axes.errorbar(pt, Ra, yerr= err, fmt = "o",color = clr, ecolor = clr, uplims=False, lolims=False, capsize = 5, capthick =0.8, barsabove = False,label=lbl)
    if(collinear == 1):
        axes.fill_between(pt_new,factor*maxxc_new,factor*minnc_new,color = clr,alpha=0.2)
```
