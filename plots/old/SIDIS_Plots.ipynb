```python
class multplot:
    
    def __init__(self, data, hadron, nucleus, var, plotcolor):
        self.data = data
        self.hadron = hadron
        self.nucleus = nucleus
        self.var = var
        self.plotcolor = plotcolor
        
    def newplot(self,label):
        plt.title(label, fontsize = 'x-large')
        plt.axhline(1,color ='black',linewidth = 0.5)
        plt.ylabel(r"\rm $R^h_A$", fontsize = 'x-large',rotation = 0,labelpad = 15)
        plt.xlabel(self.var,fontsize = 'x-large', labelpad = 10)
        
    def plot(self):
        plt.errorbar(self.data.xdata, self.data.ydata, self.data.error, fmt = "o",color = self.plotcolor, ecolor = self.plotcolor, uplims=False, lolims=False, capsize = 5, capthick =0.8, barsabove = False,label = self.nucleus)
        plt.legend(loc = 'best')
        
    def newaxesplot(self,axes,ylow,yhigh):
        axes.axhline(1,color = 'black', linewidth = 1)
        axes.tick_params(axis = 'both', direction ='in',labelsize = 'x-large')
        axes.set_ylim(ylow,yhigh)
    
    def axesplot(self,axes):
        axes.errorbar(self.data.xdata, self.data.ydata, self.data.error, fmt = "o",color = self.plotcolor, ecolor = self.plotcolor, uplims=False, lolims=False, capsize = 5, capthick =0.8, barsabove = False,label = self.nucleus)
        
    def endplot(self,ylow,yhigh):
        plt.ylim(ylow,yhigh)
        fig = plt.gcf()
        ax = plt.gca()
        ax.tick_params(axis = 'both', direction ='in',labelsize = 'x-large')
        plt.show()
        return fig, ax
        
```
