```python
class multdata:
    
    def __init__(self, xdata, ydata, staterr, systerr):
        self.xdata = xdata
        self.ydata = ydata
        self.staterr = staterr
        self.systerr = systerr
        self.error = np.sqrt(staterr**2 + systerr**2)
```

```python
class multplot2d:
    
    def __init__(self, data, bin, plotcolor):
        self.data = data
        self.bin = bin
        self.plotcolor = plotcolor
        
        
    def newaxesplot(self,axes,ylow,yhigh):
        axes.axhline(1,color = 'black', linewidth = 1)
        axes.tick_params(axis = 'both', direction ='in',labelsize = 'x-large')
        axes.set_ylim(ylow,yhigh)
    
    def axesplot(self,axes):
        axes.errorbar(self.data.xdata, self.data.ydata, self.data.error, fmt = "o",color = self.plotcolor, ecolor = self.plotcolor, uplims=False, lolims=False, capsize = 5, capthick =0.8, barsabove = False,label = self.bin)
        
        
```
