```python
class multdata:
    
    def __init__(self, file, var):
        self.xdata  = file[var]
        self.var    = var
        self.nu     = file['Nu']
        self.z      = file['Z']
        self.Q2     = file['Q2']
        self.Q      = np.sqrt(file['Q2'])
        self.pt2    = file['pt2']
        self.qt     = np.sqrt(file['pt2'])/(file['Z'])
        self.mult   = file['MULT-RATIO']
        self.stat   = file['STAT']
        self.sys    = file['SYS']
        self.multd  = file['mult_d']
        self.multa  = file['mult_a']
        self.Ra     = file['R_a']
        self.error  = np.sqrt((file['STAT'])**2+ (file['SYS'])**2)
```
